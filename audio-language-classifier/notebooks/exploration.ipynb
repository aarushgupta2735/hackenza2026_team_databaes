# Audio Language Classifier Exploration

This notebook is used for exploratory data analysis (EDA) and visualizing audio data for the audio language classification project.

```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')

# Function to load and visualize audio
def load_and_visualize_audio(file_path):
    audio, sr = librosa.load(file_path)
    plt.figure(figsize=(12, 4))
    plt.plot(audio)
    plt.title('Audio Waveform')
    plt.xlabel('Samples')
    plt.ylabel('Amplitude')
    plt.show()
    return audio, sr

# Example usage
# audio_file_path = 'path_to_audio_file.wav'
# audio, sr = load_and_visualize_audio(audio_file_path)
```

## Feature Extraction

In this section, we will extract features from the audio samples, such as MFCCs and spectrograms.

```python
# Function to extract MFCCs
def extract_mfcc(file_path, n_mfcc=13):
    audio, sr = librosa.load(file_path)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfccs

# Example usage
# mfccs = extract_mfcc(audio_file_path)
# plt.figure(figsize=(12, 4))
# plt.imshow(mfccs, aspect='auto', origin='lower')
# plt.title('MFCCs')
# plt.colorbar()
# plt.show()
```
